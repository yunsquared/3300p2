```python
import json

import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
from   collections import defaultdict
import string

```

```python
data_2013_file= 'datasets/countries-2013-updated.json'
solutions2013_file = 'datasets/solutions-2013-sorted.json'
```

```python
data2013_json = pd.read_json(data_2013_file)

solutions2013_json = pd.read_json(solutions2013_file)
```

```python
data2013_json.head()
```

```python
solutions2013_json = solutions2013_json.drop('destination_countries', axis=1)
```

```python
for country in solutions2013_json['country']:
    print(type(country))
```

```python
merged_df = pd.merge(data2013_json, solutions2013_json, on='country')
```

```python
merged_df.sample(10)
```

```python
origin_data = merged_df[merged_df['country']=="Liberia"]
print(origin_data.iloc[0]['destination_countries'])
```

```python
merged_dict=[]
for i, country in enumerate(merged_df['country']):
    origin_data = merged_df[merged_df['country']==country]
    
    merged_dict.append({"country": country, "destination_countries": [], 
                        "destination_refugees": [] , 
                        "destination_pairs": [], 
                        "totals": [] })

#     total_refugees = 0.0
#     total_asylum = 0.0
#     total_returned = 0.0
#     total_resettled = 0.0
#     total_naturalized = 0.0
    
    
    merged_dict[i]["destination_countries"] = merged_df.iloc[i]['destination_countries']
    
    merged_dict[i]["destination_refugees"] = merged_df.iloc[i]['destination_refugees']
    
    merged_dict[i]["destination_pairs"] = merged_df.iloc[i]['destination_pairs']
    
    merged_dict[i]["totals"] = float(merged_df.iloc[i]['totals'][0])
    merged_dict[i]['returned_refugees'] = float(merged_df.iloc[i]['returned_refugees'])
    
    merged_dict[i]['resettled'] = float(merged_df.iloc[i]['resettlement'])
    merged_dict[i]['naturalized'] = float(merged_df.iloc[i]['naturalized'])
    
    
print(merged_dict) 
```

```python
with open('merged_2013.json', 'w') as f: 
    json.dump(merged_dict,f)
```
